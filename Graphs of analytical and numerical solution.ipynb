```python
%matplotlib inline

from numpy import linalg as la
import numpy as np
import matplotlib.pyplot as plt

from numpy import asarray
from numpy import savetxt

from numpy import loadtxt

# inisiasi variabel posisi x
N = 100
L = 10.0
dx = L/N
x0 = L/2
x = np.arange(N+1)*dx - x0
```

### Validasi Solusi Analitik dan Numerik

Fungsi Gelombang analitik dan numerik $m_I=0$

```python
#ANALITIK
#upload data fungsi gelombang riil solusi analitik mi=0, dari n1-n4
psi_m0an0 = loadtxt('E:\scriptpsi\data\psi_ana_realm0.0n0.csv', delimiter=',')
psi_m0an1 = loadtxt('E:\scriptpsi\data\psi_ana_realm0.0n1.csv', delimiter=',')
psi_m0an2 = loadtxt('E:\scriptpsi\data\psi_ana_realm0.0n2.csv', delimiter=',')
psi_m0an3 = loadtxt('E:\scriptpsi\data\psi_ana_realm0.0n3.csv', delimiter=',')
psi_m0an4 = loadtxt('E:\scriptpsi\data\psi_ana_realm0.0n4.csv', delimiter=',')
#upload data fungsi gelombang imajiner solusi analitik mi=0, dari n1-n4
psiim_m0an0 = loadtxt('E:\scriptpsi\data\psi_ana_imagm0.0n0.csv', delimiter=',')
psiim_m0an1 = loadtxt('E:\scriptpsi\data\psi_ana_imagm0.0n1.csv', delimiter=',')
psiim_m0an2 = loadtxt('E:\scriptpsi\data\psi_ana_imagm0.0n2.csv', delimiter=',')
psiim_m0an3 = loadtxt('E:\scriptpsi\data\psi_ana_imagm0.0n3.csv', delimiter=',')
psiim_m0an4 = loadtxt('E:\scriptpsi\data\psi_ana_imagm0.0n4.csv', delimiter=',')

#NUMERIK
#upload data fungsi gelombang riil solusi numerik mi=0, dari n1 - n4
psi_m0num0 = loadtxt('E:\scriptpsi\data\psi_numerik_realm0jn0.csv', delimiter=',')
psi_m0num1 = loadtxt('E:\scriptpsi\data\psi_numerik_realm0jn1.csv', delimiter=',')
psi_m0num2 = loadtxt('E:\scriptpsi\data\psi_numerik_realm0jn2.csv', delimiter=',')
psi_m0num3 = loadtxt('E:\scriptpsi\data\psi_numerik_realm0jn3.csv', delimiter=',')
psi_m0num4 = loadtxt('E:\scriptpsi\data\psi_numerik_realm0jn4.csv', delimiter=',')
#upload data fungsi gelombang riil solusi numerik mi=0, dari n1 - n4
psiim_m0num0 = loadtxt('E:\scriptpsi\data\psi_numerik_imagm0jn0.csv', delimiter=',')
psiim_m0num1 = loadtxt('E:\scriptpsi\data\psi_numerik_imagm0jn1.csv', delimiter=',')
psiim_m0num2 = loadtxt('E:\scriptpsi\data\psi_numerik_imagm0jn2.csv', delimiter=',')
psiim_m0num3 = loadtxt('E:\scriptpsi\data\psi_numerik_imagm0jn3.csv', delimiter=',')
psiim_m0num4 = loadtxt('E:\scriptpsi\data\psi_numerik_imagm0jn4.csv', delimiter=',')
```

```python
#PENGGABUNGAN BAGIAN RIIL DAN IMAJINER, mi=0
#UNTUK MENGHITUNG PROBABILITAS
#gabung menjadi fungsi gelombang kompleks: analitik
psin_m0an0 = m = np.zeros((N+1), dtype = complex)
psin_m0an1 = m = np.zeros((N+1), dtype = complex)
psin_m0an2 = m = np.zeros((N+1), dtype = complex)
psin_m0an3 = m = np.zeros((N+1), dtype = complex)
psin_m0an4 = m = np.zeros((N+1), dtype = complex)

for i in range(N+1):
    psin_m0an0[i] = complex(psi_m0an0[i], psiim_m0an0[i])
    psin_m0an1[i] = complex(psi_m0an1[i], psiim_m0an1[i])
    psin_m0an2[i] = complex(psi_m0an2[i], psiim_m0an2[i])
    psin_m0an3[i] = complex(psi_m0an3[i], psiim_m0an3[i])
    psin_m0an4[i] = complex(psi_m0an4[i], psiim_m0an4[i])
    
#gabung menjadi fungsi gelombang kompleks: numerik
psin_m0num0 = m = np.zeros((N+1), dtype = complex)
psin_m0num1 = m = np.zeros((N+1), dtype = complex)
psin_m0num2 = m = np.zeros((N+1), dtype = complex)
psin_m0num3 = m = np.zeros((N+1), dtype = complex)
psin_m0num4 = m = np.zeros((N+1), dtype = complex)

for i in range(N+1):
    psin_m0num0[i] = complex(psi_m0num0[i], psiim_m0num0[i])
    psin_m0num1[i] = complex(psi_m0num1[i], psiim_m0num1[i])
    psin_m0num2[i] = complex(psi_m0num2[i], psiim_m0num2[i])
    psin_m0num3[i] = complex(psi_m0num3[i], psiim_m0num3[i])
    psin_m0num4[i] = complex(psi_m0num4[i], psiim_m0num4[i])
```

```python
#grafik fungsi gelombang sistem bermassa riil (mi=0) pada keadaan dasar
plt.plot(x, psi_m0an0, 'b')
plt.plot(x, psi_m0num0, 'xr', markersize=5)
plt.xticks(np.arange(-5, 6))

plt.xlabel('$x$')
plt.ylabel('$\psi(x)$')
plt.legend(["analitik", "numerik"], loc= "best", prop={'size':10})

plt.savefig('E:\scriptpsi\grafik\psin_annum_m0n00.png', dpi=500,  bbox_inches="tight")
plt.show()
```

```python
#grafik absolut error fungsi gelombang sistem bermassa riil (mi=0) pada keadaan dasar
plt.plot(x, (psi_m0an0 - psi_m0num0), 'o', markersize=3)
plt.xticks(np.arange(-5, 6))
plt.xlabel('$x$')
plt.ylabel('$\Delta\psi_0(x)$')
plt.savefig('E:\scriptpsi\grafik\delta_psim0n0.png', dpi=500,  bbox_inches="tight")
plt.show()
```

```python
#grafik fungsi gelombang sistem bermassa riil (mi=0) pada keadaan eksitasi pertama-keempat
fig, axs = plt.subplots(2, 2, figsize = (9, 6))

axs[0, 0].plot(x, psiim_m0an1, 'b')
axs[0, 0].plot(x, psi_m0num1, 'xr', markersize=4)
axs[0, 0].set_xticks(np.arange(-5, 6))
axs[0, 0].set_title("n = 1")
axs[0, 0].set_ylabel("$\psi(x)$")
axs[0, 0].legend(["analitik", "numerik"], loc= "best", prop={'size':9})

axs[0, 1].plot(x, psi_m0an2, 'b')
axs[0, 1].plot(x, psi_m0num2, 'xr', markersize=4)
axs[0, 1].set_xticks(np.arange(-5, 6))
axs[0, 1].set_title("n = 2")

axs[1, 0].plot(x, psiim_m0an3, 'b')
axs[1, 0].plot(x, psi_m0num3, 'xr', markersize=4)
axs[1, 0].set_xticks(np.arange(-5, 6))
axs[1, 0].set_title("n = 3")
axs[1, 0].set_xlabel("$x$")
axs[1, 0].set_ylabel("$\psi(x)$")

axs[1, 1].plot(x, psi_m0an4, 'b')
axs[1, 1].plot(x, -psi_m0num4, 'xr', markersize=4)
axs[1, 1].set_xticks(np.arange(-5, 6))
axs[1, 1].set_title("n = 4")
axs[1, 1].set_xlabel("$x$")

plt.savefig('E:\scriptpsi\grafik\psin_annum_m0n0.png', dpi=500,  bbox_inches="tight")
fig.tight_layout()
```

```python
#grafik absolut error fungsi gelombang massa riil (mi=0) keadaan eksitasi pertama-keempat
fig, axs = plt.subplots(2, 2, figsize = (9, 6))

axs[0, 0].plot(x, (psiim_m0an1 - psi_m0num1), 'o', markersize=3)
axs[0, 0].set_xticks(np.arange(-5, 6))
axs[0, 0].set_title("n = 1")
axs[0, 0].set_ylabel("$\Delta \psi(x)$")

axs[0, 1].plot(x, (psi_m0an2 - psi_m0num2), 'o', markersize=3)
axs[0, 1].set_xticks(np.arange(-5, 6))
axs[0, 1].set_title("n = 2")

axs[1, 0].plot(x, (psiim_m0an3 - psi_m0num3), 'o', markersize=3)
axs[1, 0].set_xticks(np.arange(-5, 6))
axs[1, 0].set_title("n = 3")
axs[1, 0].set_xlabel("$x$")
axs[1, 0].set_ylabel("$\Delta \psi(x)$")

axs[1, 1].plot(x, (psi_m0an4 - (-psi_m0num4)), 'o', markersize=3)
axs[1, 1].set_xticks(np.arange(-5, 6))
axs[1, 1].set_title("n = 4")
axs[1, 1].set_xlabel("$x$")

plt.savefig('E:\scriptpsi\grafik\delta_psim0n14.png', dpi=500,  bbox_inches="tight")
fig.tight_layout()
```

Probabilitas sistem bermassa riil

```python
#plot grafik probabilitas, mi=0
fig, axs = plt.subplots(2, 2, figsize = (9, 6))

axs[0, 0].plot(x, psin_m0an1*(np.conjugate(psin_m0an1)), 'b', linewidth=1)
axs[0, 0].plot(x, psin_m0num1*(np.conjugate(psin_m0num1)), 'rx', markersize=4)
axs[0, 0].set_xticks(np.arange(-5, 6))
axs[0, 0].set_title("n = 1")
axs[0, 0].set_ylabel("$|\psi(x)|^2$")
axs[0, 0].legend(["analitik", "numerik"], loc= "best", prop={'size':9})

axs[0, 1].plot(x, psin_m0an2*(np.conjugate(psin_m0an2)), 'b', linewidth=1)
axs[0, 1].plot(x, psin_m0num2*(np.conjugate(psin_m0num2)), 'rx', markersize=4)
axs[0, 1].set_xticks(np.arange(-5, 6))
axs[0, 1].set_title("n = 2")

axs[1, 0].plot(x, psin_m0an3*(np.conjugate(psin_m0an3)), 'b', linewidth=1)
axs[1, 0].plot(x, psin_m0num3*(np.conjugate(psin_m0num3)), 'rx', markersize=4)
axs[1, 0].set_xticks(np.arange(-5, 6))
axs[1, 0].set_title("n = 3")
axs[1, 0].set_xlabel("$x$")
axs[1, 0].set_ylabel("$|\psi(x)|^2$")

axs[1, 1].plot(x, psin_m0an4*(np.conjugate(psin_m0an4)), 'b', linewidth=1)
axs[1, 1].plot(x, psin_m0num4*(np.conjugate(psin_m0num4)), 'rx', markersize=4)
axs[1, 1].set_xticks(np.arange(-5, 6))
axs[1, 1].set_title("n = 4")
axs[1, 1].set_xlabel("$x$")

plt.savefig('E:\scriptpsi\grafik\psi_probm0n14.png', dpi=500,  bbox_inches="tight")
fig.tight_layout()
```

```python
#absolut error probabilitas, mi=0
delta_psin_m0n0 = psin_m0an0*(np.conjugate(psin_m0an0)) - psin_m0num0*(np.conjugate(psin_m0num0))
delta_psin_m0n1 = psin_m0an1*(np.conjugate(psin_m0an1)) - psin_m0num1*(np.conjugate(psin_m0num1))
delta_psin_m0n2 = psin_m0an2*(np.conjugate(psin_m0an2)) - psin_m0num2*(np.conjugate(psin_m0num2))
delta_psin_m0n3 = psin_m0an3*(np.conjugate(psin_m0an3)) - psin_m0num3*(np.conjugate(psin_m0num3))
delta_psin_m0n4 = psin_m0an4*(np.conjugate(psin_m0an4)) - psin_m0num4*(np.conjugate(psin_m0num4))


#plot grafik absolut error probabilitas, mi=0
fig, axs = plt.subplots(2, 2, figsize = (9, 6))

axs[0, 0].plot(x, delta_psin_m0n1, 'o', markersize=3)
axs[0, 0].set_xticks(np.arange(-5, 6))
axs[0, 0].set_title("n = 1")
axs[0, 0].set_ylabel("$\Delta |\psi(x)|^2$")

axs[0, 1].plot(x, delta_psin_m0n2, 'o', markersize=3)
axs[0, 1].set_xticks(np.arange(-5, 6))
axs[0, 1].set_title("n = 2")

axs[1, 0].plot(x, delta_psin_m0n3, 'o', markersize=3)
axs[1, 0].set_xticks(np.arange(-5, 6))
axs[1, 0].set_title("n = 3")
axs[1, 0].set_xlabel("$x$")
axs[1, 0].set_ylabel("$\Delta|\psi(x)|^2$")

axs[1, 1].plot(x, delta_psin_m0n4, 'o', markersize=3)
axs[1, 1].set_xticks(np.arange(-5, 6))
axs[1, 1].set_title("n = 4")
axs[1, 1].set_xlabel("$x$")

plt.savefig('E:\scriptpsi\grafik\delta_probm0n14.png', dpi=500,  bbox_inches="tight")
fig.tight_layout()
```
